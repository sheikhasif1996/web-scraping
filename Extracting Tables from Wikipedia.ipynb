```python
import pandas as pd
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_research_universities_in_the_United_States')
```

```python
c=0
for uni in data:
    if c==0:
        uni1 = uni
    elif c==1:
        uni2 = uni
    elif c==2:
        break
    c+=1
```
